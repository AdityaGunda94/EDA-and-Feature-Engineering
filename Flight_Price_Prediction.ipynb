### Flight price prediction - EDA and Feature Eng

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel('Data_Train.xlsx')
train_df.head()
```

```python
test_df = pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
final_df = train_df.append(test_df)
```

```python
final_df.head()
```

```python

final_df.tail()
```

```python
final_df.info()
```

```python
#from Date_of_Journey variable, lets derive data, month and year separetely.

final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
final_df.info()
```

```python
#convert date month and year in integer

final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
#drop the column
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
#Arrival time

#final_df['Arrival_Time'].str.split(' ').str[0]

final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
final_df['Arrival_Time']
```

```python
final_df['Arrival_hour'] = final_df['Arrival_Time'].str.split(":").str[0]
```

```python
final_df['Arrival_min'] = final_df['Arrival_Time'].str.split(":").str[1]
```

```python
final_df.head(1)
```

```python
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_min'] = final_df['Arrival_min'].astype(int)
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
#similarly for dep_time 
```

```python
final_df['Dep_hour'] = final_df['Dep_Time'].str.split(":").str[0]
final_df['Dep_min'] = final_df['Dep_Time'].str.split(":").str[1]
```

```python
final_df.head(1)
```

```python
final_df['Dep_hour'] = final_df['Dep_hour'].astype(int)
final_df['Dep_min'] = final_df['Dep_min'].astype(int)
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df.head()
```

```python
#drop the route column
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
#lets handle Duration.
# we can see that there are 2 records with only minutes mentioned. lets drop those.
```

```python
final_df[final_df['Duration']== '5m']
```

```python
final_df = final_df[~final_df['Duration']=='5m']
```

```python
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df[final_df['Duration']== '5m']
```

```python
final_df['duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df.head()
```

```python
final_df['duration_hour'] = final_df['duration_hour'].astype(int)
```

```python
final_df['duration_min']=final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df.head()
```

```python
final_df.drop('Duration',axis=1,inplace=True)
```

```python
final_df['duration_min'] = final_df['duration_min'].astype(int)
```

```python
final_df['duration_min'].unique()
```

```python
final_df['duration_min'].isnull().sum()
```

```python
#drop the records where duration_min is null
```

```python
final_df = final_df[~final_df['duration_min'].isnull()]
```

```python
final_df['duration_min'] = final_df['duration_min'].astype(int)
```

```python
final_df['Airline'].unique()
```

```python
#lets do label encoding for Airlines - we can use get dummies also. but it is not recommended here because, here we are dealing with both train and test data
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
final_df['Airline'] = labelencoder.fit_transform(final_df['Airline'])
final_df['Source'] = labelencoder.fit_transform(final_df['Source'])
final_df['Destination'] = labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = labelencoder.fit_transform(final_df['Additional_Info'])
```
